```sql
DROP TABLE IF EXISTS case_study_de.customer_bronze;
DROP TABLE IF EXISTS case_study_de.order_bronze;
CREATE TABLE IF NOT EXISTS case_study_de.customer_bronze (
  customer_nk VARCHAR(24) NOT NULL,
  first_name VARCHAR(300),
  last_name VARCHAR(300),
  phone VARCHAR (20),
  country VARCHAR(60),
  email_id VARCHAR(320),
  customer_nk_num LONG NOT NULL,
  customer_version SMALLINT NOT NULL DEFAULT 1,
  PRIMARY KEY (customer_nk_num , customer_version) NOT ENFORCED
) USING DELTA 
PARTITIONED BY (customer_version)
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported');

CREATE TABLE IF NOT EXISTS case_study_de.order_bronze (
  order_nk VARCHAR(23) NOT NULL,
  order_timestamp TIMESTAMP,
  order_date DATE GENERATED ALWAYS AS (CAST(order_timestamp AS DATE)),
  product_nk VARCHAR(23),
  customer_nk VARCHAR(24),
  customer_nk_num LONG NOT NULL,
  customer_version SMALLINT NOT NULL DEFAULT 1,
  revenue DECIMAL(10,2),
  order_status VARCHAR(10),
  order_nk_num LONG NOT NULL,
  order_version SMALLINT NOT NULL DEFAULT 1,
  product_nk_num LONG NOT NULL,
  product_version SMALLINT NOT NULL DEFAULT 1,
  PRIMARY KEY (order_nk_num , order_version) NOT ENFORCED,
  FOREIGN KEY (customer_nk_num , customer_version)
        REFERENCES case_study_de.customer_bronze(customer_nk_num , customer_version)
        NOT ENFORCED
) USING DELTA 
PARTITIONED BY (
    order_version,
    customer_version, 
    product_version,
    order_date
    )
TBLPROPERTIES ('delta.feature.allowColumnDefaults' = 'supported');
```
