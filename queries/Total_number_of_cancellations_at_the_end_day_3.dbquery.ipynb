```sql
WITH RECURSIVE dates_cte AS (
    SELECT MIN(order_date) AS incremental_date
    FROM case_study_de.order_bronze
    UNION ALL
    SELECT incremental_date + INTERVAL '1 day'
    FROM dates_cte
    WHERE incremental_date < (SELECT MAX(order_date) FROM case_study_de.order_bronze)
    ),
end_of_3 AS (SELECT incremental_date FROM dates_cte ORDER BY incremental_date LIMIT 3),
cancellations AS (SELECT * FROM case_study_de.order_bronze WHERE order_status = 'cancelled'),
cancel_day_3 AS (SELECT * FROM cancellations WHERE order_date IN (SELECT * FROM end_of_3))
SELECT COUNT(*), order_version 
FROM cancel_day_3 
GROUP BY order_version ORDER BY order_version
```
