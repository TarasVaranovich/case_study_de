```sql
-- As a updated profile definition was taken
-- profile which has orders in day 2/day 3 period
-- and has more than one version. Orders needed for time 
-- period definition. Otherwise condition has to be clarified
-- As a day 2 or day 3 was taken business definition 
-- from totl amount sum/commulative becasue no any 
-- other definition was mentioned
WITH customer_versions_on_day AS 
(SELECT DISTINCT customer_nk_num, until_day, COUNT(*) 
OVER (PARTITION BY customer_nk_num, until_day) AS versions_count
FROM case_study_de.profiles_data_days_2_3
WHERE nk_num_order IS NOT NULL 
ORDER BY customer_nk_num ASC),
updated_profile_and_day AS (SELECT DISTINCT customer_nk_num, until_day 
FROM customer_versions_on_day WHERE versions_count > 1)
SELECT upd.until_day, cb.* FROM case_study_de.customer_bronze cb
INNER JOIN updated_profile_and_day upd
ON cb.customer_nk_num = upd.customer_nk_num
ORDER BY upd.until_day, customer_nk_num, customer_version ASC
```
