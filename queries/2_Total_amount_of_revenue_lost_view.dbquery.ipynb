```sql
-- first day of the event 
-- first three days considered
-- net - sum for each day
-- commulative - increasing for each day
CREATE OR REPLACE VIEW day_amount AS 
SELECT SUM(revenue) as sum_r, order_date, order_status, order_version 
FROM case_study_de.order_bronze
GROUP BY order_date,  order_status, order_version
ORDER BY order_date,  order_status, order_version;
SELECT * FROM day_amount;
```
