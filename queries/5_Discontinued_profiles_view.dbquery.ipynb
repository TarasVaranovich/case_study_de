```sql
CREATE OR REPLACE VIEW case_study_de.sequential_date AS
WITH RECURSIVE dates_cte AS (
    SELECT MIN(order_date) AS incremental_date
    FROM case_study_de.order_bronze
    
    UNION ALL
    
    SELECT incremental_date + INTERVAL '1 day'
    FROM dates_cte
    WHERE incremental_date < (
        SELECT MAX(order_date)
        FROM case_study_de.order_bronze
    )
)
SELECT incremental_date
FROM dates_cte;
```
