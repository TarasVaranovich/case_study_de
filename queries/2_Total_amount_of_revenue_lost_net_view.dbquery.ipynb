```sql
CREATE OR REPLACE VIEW amount_lost_net AS 
SELECT * FROM (
    SELECT *
    FROM day_amount
    WHERE order_status = 'returned'
      AND order_version = 1
    ORDER BY order_date ASC
    LIMIT 3
) version_1_r
UNION ALL
SELECT * FROM (
    SELECT * 
    FROM day_amount
    WHERE order_status = 'returned'
      AND order_version = 2
    ORDER BY order_date ASC
    LIMIT 3
) version_2_r
UNION ALL
SELECT * FROM (
    SELECT * 
    FROM day_amount
    WHERE order_status = 'returned'
      AND order_version = 3
    ORDER BY order_date ASC
    LIMIT 3
) version_3_r
UNION ALL
SELECT * FROM (
    SELECT * 
    FROM day_amount
    WHERE order_status = 'cancelled'
      AND order_version = 1
    ORDER BY order_date ASC
    LIMIT 3
) version_1_c
UNION ALL
SELECT * FROM (
    SELECT * 
    FROM day_amount
    WHERE order_status = 'cancelled'
      AND order_version = 2
    ORDER BY order_date ASC
    LIMIT 3
) version_2_c
UNION ALL
SELECT * FROM (
    SELECT * 
    FROM day_amount
    WHERE order_status = 'cancelled'
      AND order_version = 3
    ORDER BY order_date ASC
    LIMIT 3
) version_3_c
ORDER BY order_version, order_status, order_date ASC;
SELECT * FROM amount_lost_net;
```
