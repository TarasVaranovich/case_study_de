```sql
SELECT  order_date, order_status, order_version, SUM(`sum(revenue)`)
OVER (ORDER BY order_date ASC) as commulative_amount
 FROM amount_lost_net 
WHERE order_status = 'returned' AND order_version = 1
UNION ALL
SELECT  order_date, order_status, order_version, SUM(`sum(revenue)`)
OVER (ORDER BY order_date ASC) as commulative_amount
 FROM amount_lost_net 
WHERE order_status = 'returned' AND order_version = 2
UNION ALL
SELECT  order_date, order_status, order_version, SUM(`sum(revenue)`)
OVER (ORDER BY order_date ASC) as commulative_amount
 FROM amount_lost_net 
WHERE order_status = 'returned' AND order_version = 3
UNION ALL
SELECT  order_date, order_status, order_version, SUM(`sum(revenue)`)
OVER (ORDER BY order_date ASC) as commulative_amount
 FROM amount_lost_net 
WHERE order_status = 'cancelled' AND order_version = 1
UNION ALL
SELECT  order_date, order_status, order_version, SUM(`sum(revenue)`)
OVER (ORDER BY order_date ASC) as commulative_amount
 FROM amount_lost_net 
WHERE order_status = 'cancelled' AND order_version = 2
UNION ALL
SELECT  order_date, order_status, order_version, SUM(`sum(revenue)`)
OVER (ORDER BY order_date ASC) as commulative_amount
 FROM amount_lost_net 
WHERE order_status = 'cancelled' AND order_version = 3
ORDER BY order_version, order_status, order_date ASC;
```
